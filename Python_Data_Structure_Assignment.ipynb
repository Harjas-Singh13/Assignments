1. Discuss String Slicing and provide examples.

 --> String Slicing in Python is a technique used to extract a portion of a string by specifying a range of indices which will help in obtaining substrings from a given string efficiently.
 The basic syntax of slicing is as follows:-
 string[start:stop:step]


*   start: The starting index of the slice(It is included)
*   stop: The ending index of the slice (exclusive)
*   step: The step size or stride between indices (optional).

Example: Extracting a Substring:

        text = "Hello World!"
        print(text[0:5])
In this example, the slice [0:5] extracts the substring from index 0 to 4.

        text = "Hello World!"
        print(text[0:-3])

In this example, the slice [0:-3] extracts the substring from the end point to the starting.


2. Explain the key features of lists in Python.
  
  --> In Python, a list is a collection of items that are ordered, changeable, and allow duplicate values. Lists are created using squared brackets and can contain items of any data type.
  
  Following are some of the key features of List in Python:-  

  a.) Ordered:- Elements maintain the order in which they are added. There is no changes even if there is a duplicate element that gets added.

  b.) Mutable: Lists can be modified, allowing for changes in elements, addition, or removal.

  c.) Dynamic Size: Lists can grow and shrink in size, allowing for flexible storage.

  d.) Heterogeneous: Lists can contain elements of different data types (e.g., integers, strings, objects).
  
  e.) Indexing and Slicing: Elements can be accessed via zero-based indexing and can be sliced to create sublists.
  
  f.) Built-in Methods: Python provides numerous built-in methods for list manipulation, such as append(), extend(), insert(), remove(), pop(), and sort().
  
  g.) Comprehensions: List comprehensions allow for concise creation and transformation of lists.
  
  h.) Nested Lists: Lists can contain other lists, enabling the creation of multi-dimensional arrays.

  i.) Iterable: Lists can be iterated over using loops, allowing easy access to each element.

  4. Compare and contrast tuples and lists with examples.

  --> i.) Mutability:-
  
  Lists: Mutable, meaning you can change, add, or remove elements after creation.

  Tuples: Immutable, meaning once created, you cannot change, add, or remove elements.

  Example:-

    #List
    my_list = [1, 2, 3]
    my_list[0] = 4            #Modifying the value
    my_list.append(5)         #Adding an element
    print(my_list)

  Output:- [4, 2, 3, 5]

    #Tuple
    my_tuple = (1, 2, 3)
    my_tuple[0] = 10  # This would raise a TypeError
    my_tuple.append(4)  # This would raise an AttributeError
    print(my_tuple)  
  
  Output:- (1, 2, 3)

  ii.) Syntax:-

  Lists: Defined using square brackets [].
  
  Tuples: Defined using parentheses ().

  Example:-

    # List
    my_list = [1, 2, 3]

    # Tuple
    my_tuple = (1, 2, 3)

  iii.) Performance

  Lists: Generally have a larger memory overhead because they are designed to accommodate dynamic changes.

  Tuples: More memory-efficient due to their immutability, which can lead to faster performance for certain operations.

  iv) Use Cases

  Lists: Suitable for collections of items that may change, such as a list of user inputs or data entries.
  
  Tuples: Ideal for fixed collections of items, such as coordinates or records that should not change.

  Example:-

    # List of user inputs
    user_inputs = ["Alice", "Bob", "Charlie"]

    # Tuple of coordinates
    coordinates = (10.0, 20.0)

  v) Methods

  Lists: Have more built-in methods for modification, such as append(), remove(), sort(), etc.
  
  Tuples: Have fewer methods, primarily count() and index().

  Example:-
    # List methods
    my_list = [1, 2, 3]
    my_list.append(4)  # Adding an element
    print(my_list)  # Output: [1, 2, 3, 4]

    # Tuple methods
    my_tuple = (1, 2, 3, 2)
    print(my_tuple.count(2))  # Output: 2
    print(my_tuple.index(3))  # Output: 2

5. Describe the key features of sets and provide examples of their use.

 --> Sets in Python are a built-in data type that store unordered collections of unique elements. Some of the key features of sets along with examples of their use:

 Key Features of Sets

  i) Unordered: Sets do not maintain the order of elements. The items are stored in a way that is optimized for membership tests.

  ii.) Unique Elements: Sets automatically remove duplicate elements. Each item can appear only once.

  iii.) Mutable: Sets are mutable, meaning you can add or remove elements after creation.

  iv) Dynamic Size: Sets can grow and shrink as elements are added or removed.

  v) Support for Mathematical Operations: Sets support various mathematical operations like union, intersection, difference, and symmetric difference.

  vi) Fast Membership Testing: Checking for the existence of an element in a set is faster than in a list, due to the underlying hash table implementation.

  Creating a Set


      # Creating a set
      my_set = {1, 2, 3, 4, 5}
      print(my_set)  # Output: {1, 2, 3, 4, 5}

  Adding Elements

      # Adding elements to a set
      my_set.add(6)
      print(my_set)  # Output: {1, 2, 3, 4, 5, 6}

  Removing Elements

      # Removing elements from a set
      my_set.remove(3)  # Raises KeyError if the element is not found
      print(my_set)  # Output: {1, 2, 4, 5, 6}

      my_set.discard(4)  # Does not raise an error if the element is not found
      print(my_set)  # Output: {1, 2, 5, 6}

6.) Discuss the use cases of tuples and sets  in python programming

Use Cases for Tuples:-

  Immutability:

    Data Integrity: Tuples are immutable, making them suitable for storing data that should not change, such as constants or fixed configurations. This can help prevent accidental modifications.
    Example: Storing configuration settings.
    CONFIG = ("localhost", 8080)

  Heterogeneous Data:

    Multiple Data Types: Tuples can hold different types of data. This makes them ideal for representing records or structured data.
    Example: A point in a 2D space.
    point = (3, 4)  # (x, y)

  Return Multiple Values:
    
    Function Returns: Functions can return multiple values using tuples, allowing easy grouping of results.
    Example:-
    def get_coordinates():
      return (10, 20)
    x, y = get_coordinates()

 Use as Dictionary Keys:

    Immutable: Since tuples are immutable, they can be used as keys in dictionaries, whereas lists cannot.
    Example:-
    locations = { (1, 2): "Point A", (3, 4): "Point B" }

 Data Unpacking:

    Convenience: Tuples support unpacking, allowing for clean assignment of multiple values.
    Example:
    a, b, c = (1, 2, 3)

Use Cases for Sets

Unique Elements:

    Removing Duplicates: Sets are ideal for eliminating duplicate entries from a collection.
    Example:-
    my_list = [1, 2, 2, 3, 4, 4, 5]
    unique_items = set(my_list)  # Output: {1, 2, 3, 4, 5}

Membership Testing:

    Fast Lookups: Sets provide efficient membership tests (O(1) on average), making them suitable for checking if an item exists in a collection.
    Example:-
    my_set = {1, 2, 3}
    print(2 in my_set)  # Output: True

Mathematical Set Operations:

    Set Theory: Sets support operations like union, intersection, and difference, making them useful for mathematical and statistical applications.
    Example:
    set_a = {1, 2, 3}
    set_b = {3, 4, 5}
    print(set_a & set_b)  # Output: {3} (Intersection)

Data Aggregation:

    Combining Data: Sets can be used to aggregate unique values from different sources.
    Example
    set1 = {"apple", "banana"}
    set2 = {"banana", "cherry"}
    combined = set1 | set2  # Output: {"apple", "banana", "cherry"}

Tracking Elements:
    
    Efficient Tracking: Sets can be used to track elements that have been seen or processed, such as user sessions or unique visitor counts.
    Example:
    visited_pages = set()
    visited_pages.add("homepage")
    visited_pages.add("about")

8. Discuss the importance of dictionary keys being immutable and provide examples.

  --> In Python, dictionary keys must be immutable. This means that the keys cannot be changed after they are created. The immutability of keys is crucial for several reasons:

  1. Hashing and Performance

    Dictionaries use a hash table to store key-value pairs. The hash function computes a unique hash value based on the key, which allows for fast data retrieval. If the key were mutable (e.g., a list), changing the key after it has been added would make it impossible to locate the value associated with that key because its hash would change.

    Example:


      # Using a tuple (immutable) as a key
      my_dict = { (1, 2): "point A", (3, 4): "point B" }
      print(my_dict[(1, 2)])  # Output: point A
      # Trying to use a list (mutable) as a key will raise an error
      # my_dict = { [1, 2]: "point A" }  # Raises TypeError

  2. Data Integrity

      Using immutable keys ensures the integrity of the data structure. Once a key is set, it cannot be changed, which prevents accidental modifications that could lead to data inconsistencies.

      Example:

          my_dict = {'name': 'Alice', 'age': 30}
          # You cannot change 'name' to something else; you can only modify the value
          my_dict['name'] = 'Bob'  # Allowed
          # But you cannot do this:
          # my_dict['name'][0] = 'C'  # Raises TypeError since strings are immutable

      
   3. Support for Hashable Types
        
        Only hashable types (like strings, numbers, and tuples) can be used as dictionary keys. This allows developers to create complex keys using immutable collections.

        Example:-

          # Tuple as a complex key
          my_dict = { (1, 2): "point A", (3, 4): "point B" }
          print(my_dict[(3, 4)])  # Output: point B
          # Using frozenset (an immutable version of a set)
          my_dict_frozen = { frozenset([1, 2]): "set A" }
          print(my_dict_frozen[frozenset([1, 2])])  # Output: set A
